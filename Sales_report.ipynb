```python
import pandas as pd
```

```python
Data = pd.read_excel(r'/content/gdrive/My Drive/workshop-day-01/Sales report.xlsx')
```

```python
display(pd.DataFrame(Data))
```

```python
Data.info()
```

```python
Data['Year'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Calculate value counts of 'Year' column
year_counts = Data['Year'].value_counts()

# Plotting using matplotlib
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Years in Sales Report Data')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group by 'Year' and calculate sum of 'SALES' for each year
yearly_sales = Data.groupby('Year')['SALES'].sum().reset_index()

# Display the yearly sales data
print("Year-wise Sales Data:")
print(yearly_sales)
```

```python
# Plotting the yearly sales as a line chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales['Year'], yearly_sales['SALES'], marker='o', color='skyblue', linestyle='-', linewidth=2)
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(yearly_sales['Year'])  # Set x-axis ticks to the years in the DataFrame
plt.show()
```

```python
country_sales = Data.groupby('COUNTRY')['SALES'].sum().reset_index()
```

```python
# Plotting country-wise sales as a bar chart
plt.figure(figsize=(12, 8))
plt.bar(country_sales['COUNTRY'], country_sales['SALES'], color='skyblue')
plt.title('Country-wise Sales Distribution')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
product_sales = Data.groupby('Product')['SALES'].sum().reset_index()
product_sales
```

```python
# Plotting country-wise sales as a bar chart
plt.figure(figsize=(12, 8))
plt.bar(product_sales['Product'], product_sales['SALES'], color='skyblue')
plt.title('Product-wise Sales Distribution')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Aggregate data by 'Product' and calculate sum of 'SALES', 'Tax', and 'Profit'
product_summary = Data.groupby('Product')[['SALES', 'Tax', 'Profit']].sum().reset_index()
product_summary
```

```python
# Plotting stacked bar chart for product-wise sales, tax, and profit
plt.figure(figsize=(12, 8))
plt.bar(product_summary['Product'], product_summary['SALES'], color='skyblue', label='Sales')
plt.bar(product_summary['Product'], product_summary['Tax'], color='orange', label='Tax', bottom=product_summary['SALES'])
plt.bar(product_summary['Product'], product_summary['Profit'], color='green', label='Profit', bottom=product_summary['SALES'] + product_summary['Tax'])
plt.title('Product-wise Sales, Tax, and Profit')
plt.xlabel('Product')
plt.ylabel('Amount')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.show()
```
